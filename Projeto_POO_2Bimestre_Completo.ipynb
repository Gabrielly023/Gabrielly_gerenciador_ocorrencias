```python
# Importa a classe datetime para manipula√ß√£o de datas
from datetime import datetime

# Importa o m√≥dulo mysql.connector para conectar ao banco de dados MySQL
import mysql.connector

# Fun√ß√£o que realiza a conex√£o com o banco de dados
def conectar_banco():
    try:
        # Estabelece conex√£o com o banco de dados MySQL local
        conexao = mysql.connector.connect(
            host="localhost",       # Endere√ßo do servidor do banco
            user="root",            # Nome de usu√°rio do banco
            password="",            # Senha do banco (vazia aqui)
            database="registro_ocorrencias"  # Nome do banco de dados
        )
        print("Conex√£o com o banco de dados estabelecida com sucesso.")
        return conexao  # Retorna o objeto de conex√£o
    except mysql.connector.Error as err:
        # Em caso de erro, exibe a mensagem e retorna None
        print(f"Erro ao conectar ao banco de dados: {err}")
        return None

# Classe base para representar ocorr√™ncias escolares simples
class OcorrenciaEscolar:
    def __init__(self, aluno, data, descricao):
        self.aluno = aluno            # Nome do aluno
        self.data = data              # Data da ocorr√™ncia
        self._descricao = descricao   # Descri√ß√£o da ocorr√™ncia (privada)
        self.id = None                # ID da ocorr√™ncia no banco (definido ap√≥s salvar)

    # M√©todo para exibir os dados da ocorr√™ncia
    def exibir_ocorrencia(self):
        print(f"\nAluno: {self.aluno}")
        print(f"Data: {self.data}")
        print(f"Descri√ß√£o: {self._descricao}")

    # M√©todo getter para a descri√ß√£o
    def get_descricao(self):
        return self._descricao

    # M√©todo setter para atualizar a descri√ß√£o
    def set_descricao(self, nova_descricao):
        self._descricao = nova_descricao
        print("Descri√ß√£o atualizada com sucesso!")

    # Atualiza a descri√ß√£o usando o setter
    def atualizar_descricao(self, nova_descricao):
        self.set_descricao(nova_descricao)

    # Altera a data da ocorr√™ncia
    def alterar_data(self, nova_data):
        self.data = nova_data
        print("Data da ocorr√™ncia alterada com sucesso!")

    # Verifica se h√° reincid√™ncia na descri√ß√£o da ocorr√™ncia
    def verificar_reincidencia(self):
        palavras_chave = ["segunda vez", "terceira vez", "reincidente"]
        descricao_lower = self._descricao.lower()
        for palavra in palavras_chave:
            if palavra in descricao_lower:
                print("‚ö†Ô∏è Aluno reincidente em comportamento inadequado.")
                return True
        print("Sem sinais de reincid√™ncia registrados.")
        return False

    # Salva a ocorr√™ncia no banco de dados
    def salvar_no_banco(self, conexao):
        try:
            cursor = conexao.cursor()
            # Query SQL para inserir os dados
            query = """
                INSERT INTO ocorrencias (aluno, data, descricao, tipo)
                VALUES (%s, %s, %s, %s)
            """
            # Converte a data de string para objeto date
            data_formatada = datetime.strptime(self.data, "%d/%m/%Y").date()
            # Executa a query com os valores
            cursor.execute(query, (
                self.aluno,
                data_formatada,
                self.get_descricao(),
                'simples'
            ))
            conexao.commit()             # Confirma a transa√ß√£o
            self.id = cursor.lastrowid   # Armazena o ID gerado
            print("‚úÖ Ocorr√™ncia simples registrada no banco de dados.")
        except mysql.connector.Error as err:
            print(f"Erro ao registrar ocorr√™ncia: {err}")
        finally:
            cursor.close()  # Fecha o cursor mesmo em caso de erro

    # Exclui a ocorr√™ncia do banco de dados
    def excluir_no_banco(self, conexao):
        try:
            if self.id is None:
                print("Erro: ID da ocorr√™ncia n√£o encontrado.")
                return
            cursor = conexao.cursor()
            # Query para deletar ocorr√™ncia com base no ID
            query = "DELETE FROM ocorrencias WHERE id = %s"
            cursor.execute(query, (self.id,))
            conexao.commit()
            print(f"‚úÖ Ocorr√™ncia com ID {self.id} exclu√≠da com sucesso.")
        except mysql.connector.Error as err:
            print(f"Erro ao excluir ocorr√™ncia: {err}")
        finally:
            cursor.close()

# Classe que representa uma ocorr√™ncia grave (herda da classe OcorrenciaEscolar)
class OcorrenciaGrave(OcorrenciaEscolar):
    def __init__(self, aluno, data, descricao, nivel_gravidade):
        super().__init__(aluno, data, descricao)  # Chama o construtor da classe base
        self._nivel_gravidade = nivel_gravidade   # N√≠vel de gravidade (privado)

    # Getter do n√≠vel de gravidade
    def get_nivel_gravidade(self):
        return self._nivel_gravidade

    # Setter do n√≠vel de gravidade
    def set_nivel_gravidade(self, novo_nivel):
        self._nivel_gravidade = novo_nivel
        print("N√≠vel de gravidade atualizado com sucesso!")

    # Exibe os dados da ocorr√™ncia grave
    def exibir_ocorrencia(self):
        print("\n‚ö†Ô∏è ALERTA DE OCORR√äNCIA GRAVE ‚ö†Ô∏è")
        print(f"Aluno: {self.aluno}")
        print(f"Data: {self.data}")
        print(f"Descri√ß√£o: {self.get_descricao()}")
        print(f"N√≠vel de Gravidade: {self._nivel_gravidade}")

    # Salva a ocorr√™ncia grave em duas tabelas: ocorrencias e ocorrencias_graves
    def salvar_no_banco(self, conexao):
        try:
            cursor = conexao.cursor()
            # Insere na tabela principal
            query_ocorrencia = """
                INSERT INTO ocorrencias (aluno, data, descricao, tipo)
                VALUES (%s, %s, %s, %s)
            """
            data_formatada = datetime.strptime(self.data, "%d/%m/%Y").date()
            cursor.execute(query_ocorrencia, (
                self.aluno,
                data_formatada,
                self.get_descricao(),
                'grave'
            ))
            conexao.commit()
            self.id = cursor.lastrowid

            # Insere o n√≠vel de gravidade na tabela espec√≠fica
            query_grave = """
                INSERT INTO ocorrencias_graves (id, nivel_gravidade)
                VALUES (%s, %s)
            """
            cursor.execute(query_grave, (self.id, self._nivel_gravidade))
            conexao.commit()

            print("‚úÖ Ocorr√™ncia grave registrada no banco de dados.")
        except mysql.connector.Error as err:
            print(f"Erro ao registrar ocorr√™ncia grave: {err}")
        finally:
            cursor.close()

# Fun√ß√£o principal que executa o sistema de registro
def rodar_sistema():
    conexao = conectar_banco()  # Conecta ao banco de dados
    if not conexao:
        print("N√£o foi poss√≠vel conectar ao banco de dados. Sistema encerrado.")
        return

    while True:
        # Interface interativa com o usu√°rio
        print("\n=== Registro de Ocorr√™ncia ===")
        aluno = input("Nome do aluno: ").strip()
        data = input("Data da ocorr√™ncia (dd/mm/aaaa): ").strip()
        descricao = input("Descri√ß√£o da ocorr√™ncia: ").strip()
        tipo = input("A ocorr√™ncia √© grave? (s/n): ").strip().lower()

        # Cria objeto do tipo apropriado
        if tipo == 's':
            nivel = input("N√≠vel de gravidade (Alta/M√©dia/Baixa): ").strip().capitalize()
            ocorrencia = OcorrenciaGrave(aluno, data, descricao, nivel)
        else:
            ocorrencia = OcorrenciaEscolar(aluno, data, descricao)

        ocorrencia.exibir_ocorrencia()         # Exibe os dados
        ocorrencia.verificar_reincidencia()    # Verifica reincid√™ncia

        # Permite ao usu√°rio atualizar a descri√ß√£o
        if input("\nDeseja atualizar a descri√ß√£o? (s/n): ").strip().lower() == 's':
            nova_desc = input("Digite a nova descri√ß√£o: ")
            ocorrencia.atualizar_descricao(nova_desc)

        # Permite ao usu√°rio alterar a data
        if input("Deseja alterar a data? (s/n): ").strip().lower() == 's':
            nova_data = input("Digite a nova data (dd/mm/aaaa): ")
            ocorrencia.alterar_data(nova_data)

        ocorrencia.exibir_ocorrencia()         # Exibe novamente ap√≥s altera√ß√µes
        ocorrencia.salvar_no_banco(conexao)    # Salva no banco de dados

        # Pergunta se deseja excluir a ocorr√™ncia rec√©m-criada
        if input("\nDeseja excluir esta ocorr√™ncia? (s/n): ").strip().lower() == 's':
            ocorrencia.excluir_no_banco(conexao)

        # Pergunta se deseja registrar outra ocorr√™ncia
        if input("\nDeseja registrar outra ocorr√™ncia? (s/n): ").strip().lower() != 's':
            print("\nEncerrando o sistema de ocorr√™ncias.")
            break

    conexao.close()  # Fecha a conex√£o com o banco

# Inicia o sistema quando o script √© executado
rodar_sistema()

```
