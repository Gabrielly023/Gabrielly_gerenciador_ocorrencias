## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3 
class OcorrenciaEscolar_gabrielly:
    def __init__(self, aluno, data, descricao):
        self.aluno = aluno            
        self.data = data              
        self.__descricao = descricao 

    def exibir_ocorrencia_gabrielly(self):
        print(f"\nAluno: {self.aluno}")
        print(f"Data: {self.data}")
        print(f"Descri√ß√£o: {self.__descricao}")

    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        self.__descricao = nova_descricao
        print("Descri√ß√£o atualizada com sucesso!")

    def atualizar_descricao(self, nova_descricao):
        self.set_descricao(nova_descricao)

    def alterar_data(self, nova_data):
        self.data = nova_data
        print("Data da ocorr√™ncia alterada com sucesso!")

    def verificar_reincidencia(self):
        palavras_chave = ["segunda vez", "terceira vez", "reincidente"]
        descricao_lower = self.__descricao.lower()
        for palavra in palavras_chave:
            if palavra in descricao_lower:
                print("‚ö†Ô∏è Aluno reincidente em comportamento inadequado.")
                return True
        print("Sem sinais de reincid√™ncia registrados.")
        return False


class OcorrenciaGrave(OcorrenciaEscolar_gabrielly):
    def __init__(self, aluno, data, descricao, nivel_gravidade):
        super().__init__(aluno, data, descricao)
        self.__nivel_gravidade = nivel_gravidade

    def get_nivel_gravidade(self):
        return self.__nivel_gravidade

    def set_nivel_gravidade(self, novo_nivel):
        self.__nivel_gravidade = novo_nivel
        print("N√≠vel de gravidade atualizado com sucesso!")

    def exibir_alerta(self):
        print("\n‚ö†Ô∏è ALERTA DE OCORR√äNCIA GRAVE ‚ö†Ô∏è")
        self.exibir_ocorrencia_gabrielly()
        print(f"N√≠vel de Gravidade: {self.__nivel_gravidade}")


while True:
    print("\n=== Registro de Ocorr√™ncia ===")

    aluno = input("Nome do aluno: ")
    data = input("Data da ocorr√™ncia (dd/mm/aaaa): ")
    descricao = input("Descri√ß√£o da ocorr√™ncia: ")
    tipo = input("A ocorr√™ncia √© grave? (s/n): ").strip().lower()

    if tipo == 's':
        nivel = input("N√≠vel de gravidade (Alta/M√©dia/Baixa): ")
        ocorrencia = OcorrenciaGrave(aluno, data, descricao, nivel)
        ocorrencia.exibir_alerta()
    else:
        ocorrencia = OcorrenciaEscolar_gabrielly(aluno, data, descricao)
        ocorrencia.exibir_ocorrencia_gabrielly()

    ocorrencia.verificar_reincidencia()

    atualizar_desc = input("\nDeseja atualizar a descri√ß√£o? (s/n): ").strip().lower()
    if atualizar_desc == 's':
        nova_desc = input("Digite a nova descri√ß√£o: ")
        ocorrencia.atualizar_descricao(nova_desc)

    atualizar_data = input("Deseja alterar a data? (s/n): ").strip().lower()
    if atualizar_data == 's':
        nova_data = input("Digite a nova data (dd/mm/aaaa): ")
        ocorrencia.alterar_data(nova_data)

    if tipo == 's':
        ocorrencia.exibir_alerta()
    else:
        ocorrencia.exibir_ocorrencia_gabrielly()

    continuar = input("\nDeseja registrar outra ocorr√™ncia? (s/n): ").strip().lower()
    if continuar != 's':
        print("\nEncerrando o sistema de ocorr√™ncias.")
        break

```

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4
from datetime import datetime
import mysql.connector

def conectar_banco():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="registro_ocorrencias"
    )

class OcorrenciaEscolar_gabrielly:
    def __init__(self, aluno, data, descricao):
        self.aluno = aluno
        self.data = data
        self.__descricao = descricao

    def exibir_ocorrencia_gabrielly(self):
        print(f"\nAluno: {self.aluno}")
        print(f"Data: {self.data}")
        print(f"Descri√ß√£o: {self.__descricao}")

    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        self.__descricao = nova_descricao
        print("Descri√ß√£o atualizada com sucesso!")

    def atualizar_descricao(self, nova_descricao):
        self.set_descricao(nova_descricao)

    def alterar_data(self, nova_data):
        self.data = nova_data
        print("Data da ocorr√™ncia alterada com sucesso!")

    def verificar_reincidencia(self):
        palavras_chave = ["segunda vez", "terceira vez", "reincidente"]
        descricao_lower = self.__descricao.lower()
        for palavra in palavras_chave:
            if palavra in descricao_lower:
                print("‚ö†Ô∏è Aluno reincidente em comportamento inadequado.")
                return True
        print("Sem sinais de reincid√™ncia registrados.")
        return False

    def salvar_no_banco(self, conexao):
        cursor = conexao.cursor()
        query = """
            INSERT INTO ocorrencias (aluno, data, descricao, tipo)
            VALUES (%s, %s, %s, %s)
        """
        cursor.execute(query, (
            self.aluno,
            datetime.strptime(self.data, "%d/%m/%Y").date(),
            self.get_descricao(),
            'simples'
        ))
        conexao.commit()
        print("‚úÖ Ocorr√™ncia simples registrada no banco de dados.")
        cursor.close()

class OcorrenciaGrave(OcorrenciaEscolar_gabrielly):
    def __init__(self, aluno, data, descricao, nivel_gravidade):
        super().__init__(aluno, data, descricao)
        self.__nivel_gravidade = nivel_gravidade

    def get_nivel_gravidade(self):
        return self.__nivel_gravidade

    def set_nivel_gravidade(self, novo_nivel):
        self.__nivel_gravidade = novo_nivel
        print("N√≠vel de gravidade atualizado com sucesso!")

    def exibir_ocorrencia_gabrielly(self):
        print("\n‚ö†Ô∏è ALERTA DE OCORR√äNCIA GRAVE ‚ö†Ô∏è")
        print(f"Aluno: {self.aluno}")
        print(f"Data: {self.data}")
        print(f"Descri√ß√£o: {self.get_descricao()}")
        print(f"N√≠vel de Gravidade: {self.__nivel_gravidade}")

    def salvar_no_banco(self, conexao):
        cursor = conexao.cursor()

        query_ocorrencia = """
            INSERT INTO ocorrencias (aluno, data, descricao, tipo)
            VALUES (%s, %s, %s, %s)
        """
        cursor.execute(query_ocorrencia, (
            self.aluno,
            datetime.strptime(self.data, "%d/%m/%Y").date(),
            self.get_descricao(),
            'grave'
        ))
        conexao.commit()
        ocorrencia_id = cursor.lastrowid

        query_grave = """
            INSERT INTO ocorrencias_graves (id, nivel_gravidade)
            VALUES (%s, %s)
        """
        cursor.execute(query_grave, (ocorrencia_id, self.__nivel_gravidade))
        conexao.commit()

        print("‚úÖ Ocorr√™ncia grave registrada no banco de dados.")
        cursor.close()

conexao = conectar_banco()

while True:
    print("\n=== Registro de Ocorr√™ncia ===")

    aluno = input("Nome do aluno: ")
    data = input("Data da ocorr√™ncia (dd/mm/aaaa): ")
    descricao = input("Descri√ß√£o da ocorr√™ncia: ")
    tipo = input("A ocorr√™ncia √© grave? (s/n): ").strip().lower()

    if tipo == 's':
        nivel = input("N√≠vel de gravidade (Alta/M√©dia/Baixa): ")
        ocorrencia = OcorrenciaGrave(aluno, data, descricao, nivel)
    else:
        ocorrencia = OcorrenciaEscolar_gabrielly(aluno, data, descricao)

    ocorrencia.exibir_ocorrencia_gabrielly()
    ocorrencia.verificar_reincidencia()

    atualizar_desc = input("\nDeseja atualizar a descri√ß√£o? (s/n): ").strip().lower()
    if atualizar_desc == 's':
        nova_desc = input("Digite a nova descri√ß√£o: ")
        ocorrencia.atualizar_descricao(nova_desc)

    atualizar_data = input("Deseja alterar a data? (s/n): ").strip().lower()
    if atualizar_data == 's':
        nova_data = input("Digite a nova data (dd/mm/aaaa): ")
        ocorrencia.alterar_data(nova_data)

    ocorrencia.exibir_ocorrencia_gabrielly()

    ocorrencia.salvar_no_banco(conexao)

    continuar = input("\nDeseja registrar outra ocorr√™ncia? (s/n): ").strip().lower()
    if continuar != 's':
        print("\nEncerrando o sistema de ocorr√™ncias.")
        break

conexao.close()
```
